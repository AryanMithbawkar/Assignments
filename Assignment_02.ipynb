```python
#5. Check for Missing values and deal with them.
data.isnull().value_counts()
```

```python
data['culmen_length_mm'].fillna(data['culmen_length_mm'].mean(),inplace=True)
data['culmen_depth_mm'].fillna(data['culmen_depth_mm'].mean(),inplace=True)
data['flipper_length_mm'].fillna(data['flipper_length_mm'].mean(),inplace=True)
data['body_mass_g'].fillna(data['body_mass_g'].mean(),inplace=True)
data['sex'].fillna(data['sex'].mode()[0],inplace=True)
```

```python
data.isnull().value_counts()
```

```python
#6. Find the outliers and replace them outliers
sns.set_palette('muted')
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
sns.boxplot(data=data,y='culmen_length_mm')
plt.ylabel("culmen length in mm")
plt.subplot(2,2,2)
sns.boxplot(data=data,y='culmen_depth_mm')
plt.ylabel("culmen depth in mm")
plt.tight_layout(pad=2.0)
plt.subplot(2,2,3)
sns.boxplot(data=data,y='flipper_length_mm')
plt.ylabel("flipper length in mm")
plt.subplot(2,2,4)
sns.boxplot(data=data,y='body_mass_g')
plt.ylabel("body mass in grams")
```

```python
#7. Check for Categorical columns and perform encoding.
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['species']=le.fit_transform(data['species'])
data['island']=le.fit_transform(data['island'])
data['sex']=le.fit_transform(data['sex'])
```

```python
data.info()
```

```python
#9. Split the data into dependent and independent variables.
x=data.iloc[:,1:6]
x.head(2)
```

```python
y=data.iloc[:,:1]
y.head(2)
```

```python
#10. Scaling the data
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
#11. Split the data into training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#12. Build the Model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr=LinearRegression()
r=Ridge()
l=Lasso()
dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
```

```python
#13. Train the Model
lr.fit(x_train,y_train)
```

```python
l.fit(x_train,y_train)
```

```python
r.fit(x_train,y_train)
```

```python
dt.fit(x_train,y_train)
```

```python
rf.fit(x_train,y_train)
```

```python
#14. Test the Model
pred1=lr.predict(x_test)
pred2=l.predict(x_test)
pred3=r.predict(x_test)
pred4=dt.predict(x_test)
pred5=rf.predict(x_test)
```

```python
print(pred1)
print(pred2)
print(pred3)
print(pred4)
print(pred5)
```

```python
#15. Measure the performance using Evaluation Metrics.
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error,classification_report
```

```python
print(f"for linear regression model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred1)}")
print(f"\nfor lasso model \nMSE={mean_squared_error(y_test,pred2)} \nR2_score={r2_score(y_test,pred2)}")
print(f"\nfor ridge model \nMSE={mean_squared_error(y_test,pred3)} \nR2_score={r2_score(y_test,pred3)}")
print(f"\nfor Decision tree model \naccuracy score={accuracy_score(y_test,pred4)} \nR2_score={r2_score(y_test,pred4)}")
print(f"\nfor Random tree model \naccuracy score={accuracy_score(y_test,pred5)} \nR2_score={r2_score(y_test,pred5)}")
```
